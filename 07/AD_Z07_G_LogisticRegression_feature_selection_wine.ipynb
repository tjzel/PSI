```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

# LogisticRegression feature selection

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
wine = load_wine()
print(list(wine.target_names))
print(wine.data)
wine.target[ wine.target ==0 ] = 1 # we use only two classes 
print(wine.target)
```

```python
X_train, X_test, y_train, y_test = train_test_split( wine.data, wine.target, stratify = wine.target, random_state=42)
print(X_train.shape)
```

# Zad

Wykonaj algorytm LogisticRegression z 

* C = 1


```python
from sklearn.linear_model import LogisticRegression
log_reg_1 = LogisticRegression(C = 1, random_state=42, max_iter=4096)
log_reg_1.fit(X_train, y_train)

y_proba_1 = log_reg_1.predict_proba(X_test)
```

# Zad

Wykonaj algorytm LogisticRegression z 

* C = 100


```python
log_reg_100 = LogisticRegression(C = 100, random_state=42, max_iter=4096)
log_reg_100.fit(X_train, y_train)

y_proba_100 = log_reg_100.predict_proba(X_test)
```

# Zad

Wykonaj algorytm LogisticRegression z 

* C = 0.01


```python
log_reg_001 = LogisticRegression(C = 0.01, random_state=42, max_iter=4096)
log_reg_001.fit(X_train, y_train)

y_proba_001 = log_reg_001.predict_proba(X_test)
```

# Zad

Na koniec spójrzmy na współrzędne wykorzystywane przez modele z trzema różnymi ustawieniami parametru regularyzacji C

```python
plt.figure(figsize=(10, 5))
plt.plot(log_reg_1.coef_.T, 'o', label="C=1")

plt.show()

plt.figure(figsize=(10, 5))
plt.plot(log_reg_100.coef_.T, 'o', label="C=100")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(log_reg_001.coef_.T, 'o', label="C=0.01")
plt.show()
```

# Zad

Narysuj wykre współrzędnych wykorzystywane przez modele z trzema różnymi ustawieniami parametru regularyzacji C i regularyzacją L1.

```python
log_regL1_1 = LogisticRegression(C = 1, random_state=42, max_iter=4096, penalty='l1', solver='liblinear')
log_regL1_1.fit(X_train, y_train)
y_probaL1_1 = log_regL1_1.predict_proba(X_test)

log_regL1_100 = LogisticRegression(C = 100, random_state=42, max_iter=4096, penalty='l1', solver='liblinear')
log_regL1_100.fit(X_train, y_train)
y_probaL1_100 = log_regL1_100.predict_proba(X_test)

log_regL1_001 = LogisticRegression(C = 0.01, random_state=42, max_iter=4096, penalty='l1', solver='liblinear')
log_regL1_001.fit(X_train, y_train)
y_probaL1_001 = log_regL1_001.predict_proba(X_test)

plt.figure(figsize=(10, 5))
plt.plot(log_regL1_1.coef_.T, 'o', label="C=1")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(log_regL1_100.coef_.T, 'o', label="C=100")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(log_regL1_001.coef_.T, 'o', label="C=0.01")
plt.show()
```
