````python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    import matplotlib
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
````

````python
plt.figure(figsize=(9,9))
example_images = np.r_[X[:12000:600], X[13000:30600:600], X[30600:60000:590]]
plot_digits(example_images, images_per_row=10)
plt.show()
````

````python
from sklearn.base import BaseEstimator
class Never5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
````

# Confusion Matrix

 * Aby obliczy **Confusion Matrix**, najpierw trzeba mie zestaw prognoz, aby mona je byo porwna z rzeczywistymi etykietami.

 * Moesz przewidywa na zestawie testowym (test set), ale na razie trzymaj go nietknity (pamitaj, e chcesz uy zestawu testowego tylko na samym kocu projektu, gdy masz klasyfikator, ktry jeste gotowy do uruchomienia). 
 * Zamiast tego moesz uy funkcji 
 ```pyrthon
 cross_val_predict()
 ```
 
 * Podobnie jak funkcja **cross_val_score()** funkcja **cross_val_predict()** wykonuje **K-fold cross-validation**, ale zamiast zwraca wyniki oceny, zwraca predykcje wykonane dla kadego testu. 

 * Oznacza to, e uzyskujesz czyst prognoz dla kadej instancji w zbiorze treningowym ("czysta" oznacza, e przewidywanie jest dokonywane przez model, ktry nigdy nie widzia danych podczas treningu). 

````python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(clf, X_train, y_train_5, cv=3)
y_train_pred
````

# Confusion matrix
Do **confusion_matrix()** wystarczy przekaza etykiety docelowe (y_train_5) i przewidywane klasy (y_train_pred):

````python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_train_5, y_train_pred)
````

* Kady rzd **confusion_matrix** reprezentuje rzeczywist klas, podczas gdy kada kolumna reprezentuje przewidywan klas.
* Pierwszy rzd tej macierzy uwzgldnia obrazy inne ni 5 (klasa negatywna): 
  * 53 272 z nich zostao poprawnie sklasyfikowanych jako non-5 (**true negatives**), 
  *  1 307 zostao bdnie sklasyfikowane jako 5s (**false positives**). 
* Drugi rzd uwzgldnia obrazy 5s (klasa pozytywna): 
  * 1 077 zostay bdnie sklasyfikowane jako nie-5 (**false negatives**), 
  * 4 344 zostay poprawnie sklasyfikowane jako 5s (**true positives**). 

Idealny klasyfikator miaby tylko **true positives** i **true negatives**, wic jego **confusion_matrix** miaaby niezerowe wartoci tylko na swojej gwnej przektnej (od lewej grnej do prawej dolnej):

````python
y_train_perfect_predictions = y_train_5
confusion_matrix(y_train_5, y_train_perfect_predictions)
````

**Confusion matrix** dostarcza wielu informacji, ale bardziej przydatne sa miary numeryczne. 

 * **precision** to dokadno pozytywnych przewidywa -- precyzja klasyfikatora:

$$
precision = \frac{TP}{TP + FP}
$$
 
 * Prostym sposobem na uzyskanie doskonaej precyzji jest wykonanie jednej pozytywnej prognozy i zapewnienie jej poprawnoci (precision = 1/1 = 100%). 
 * Nie byoby to zbyt uyteczne, poniewa klasyfikator zignorowaby wszystkie oprcz jednego pozytywnego wystpienia. 
 * Precyzj zwykle stosuje si wraz z inn miar o nazwie **recall**, zwan rwnie czuoci **sensitivity**.
 * **recall** jest to stosunek pozytywnych instancji, ktre s poprawnie wykrywane przez klasyfikator 
 
$$
recall = \frac{TP}{TP + FN}
$$

````python
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_train_5, y_train_pred)) 
print(recall_score(y_train_5, y_train_pred)) 
````

### Teraz widzimy, e detektor pitek nie wyglda tak dobrze. 

* Kiedy twierdzi, e obraz reprezentuje 5, jest to poprawne tylko w 79% przypadkw. 
* Co wicej, wykrywa tylko 81% z 5s.

# F1 score


 * Czsto wygodnie jest czy **precision** i **recall** w pojedyncz metryk zwan **F1 score** w szczeglnoci jeli potrzebujesz prostego sposobu na porwnanie dwch klasyfikatorw. 

 * Wynik **F1  score** jest redni harmoniczn  **precision** i **recall**. 
 
$$
F_1 = \frac{2}{ \frac{1}{precision} + \frac{1}{recall} } = 2\frac{precision * recall}{precision + recall}
= \frac{TP}{TP + \frac{FN+FP}{2}}
$$

 * Podczas gdy klasyczna rodek traktuje wszystkie wartoci w rwnym stopniu **rednia harmoniczna** nadaje znacznie wiksz wag mniejszym wartoci. 

 * W rezultacie klasyfikator uzyska wysoki wynik **F1 score** jeli zarwno  **precision** i **recall** bd wysokie.

````python
from sklearn.metrics import f1_score
f1_score(y_train_5, y_train_pred)
````

````python
from sklearn.metrics import classification_report
print(classification_report(y_train_5, y_train_pred))#, target_names=["not 5", "5"]))
````
